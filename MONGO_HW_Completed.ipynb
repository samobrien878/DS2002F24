#3

```python
rate_genre = collection.aggregate([
    {"$unwind": "$genres"},
    {"$group": {"_id": "$genres", "average_rating": {"$avg": "$imdb.rating"}}},
    {"$sort": {"average_rating": -1}},
    {"$limit": 5}
])
for i in rate_genre:
  pprint.pprint(i)
```

```python
top_directors = collection.aggregate([
{"$group": {"_id": "$directors", "avgerage_rating": {"$avg": "$imdb.rating"}}},
{"$sort": {"avgerage_rating": -1}},
{"$limit": 5}
])

for i in top_directors:
  pprint.pprint(i)
```

```python
movie_years = collection.aggregate([
    {"$group": {"_id": "$year", "count": {"$sum": 1}}},
    {"$sort": {"count": -1}},
])
for i in movie_years:
  pprint.pprint(i)
```
